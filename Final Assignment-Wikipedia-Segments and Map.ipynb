```python
table = soup.find('table', attrs={'class':'wikitable sortable'})
table_rows = table.find_all('tr')
```

```python
res = []
for tr in table_rows:
    td = tr.find_all('td')
    #title=tr.find('title')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    
    if row:
        res.append(row)


df = pd.DataFrame(res, columns=["Postcode", "Borough", "Neighborhood"])
df
```

```python
df=df[~df.Borough.str.contains('Not assigned')]
df
```

```python
df_geo=pd.read_csv('Geospatial_Coordinates.csv')
df_geo
```

```python
df_merge=pd.merge(df, df_geo, how = 'left', left_on = 'Postcode', right_on = 'Postal Code')
df_merge
```

```python
df_Toronto=df_merge[df_merge.Borough.str.contains('Toronto')]
df_Toronto
```

```python
Toronto_onehot = pd.get_dummies(df_Toronto[['Borough']], prefix="", prefix_sep="")
Toronto_onehot
```

```python
Toronto_onehot['Neighborhood'] = df_Toronto['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

Toronto_onehot.head()
```

```python
Toronto_onehot.shape
```

```python
Toronto_grouped = Toronto_onehot.groupby('Neighborhood').mean().reset_index()
Toronto_grouped
```

```python
kclusters = 3

Toronto_grouped_clustering = Toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:74] 
```

```python
# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Toronto_grouped_clustering['Neighborhood']=Toronto_grouped['Neighborhood']
df_Toronto_cluster = Toronto_grouped_clustering.join(df_Toronto.set_index('Neighborhood'), on='Neighborhood')

df_Toronto_cluster.head() # check the last columns!
```

```python
import numpy as np # library to handle data in a vectorized manner
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_Toronto_cluster['Latitude'], df_Toronto_cluster['Longitude'], df_Toronto_cluster['Neighborhood'], df_Toronto_cluster['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
