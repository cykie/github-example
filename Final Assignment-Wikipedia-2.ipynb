```python
res = []
for tr in table_rows:
    td = tr.find_all('td')
    #title=tr.find('title')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    
    if row:
        res.append(row)


df = pd.DataFrame(res, columns=["Postcode", "Borough", "Neighborhood"])
df
```
